```python
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord
import os
import pandas as pd
import numpy as np
```

```python
def cal_similarity(seq_i, seq_j):
    n_same = 0
    for l in range(len(seq_j)):
        if seq_i[l] == seq_j[l]:
            n_same = n_same +1
    sim = round(n_same/len(seq_i)*100,2) 
    return(sim)  
```

```python
file_in ='/home/mlk442/FIND/mlk442/cd_hit_results/rpob_all_muscle_trim-2.fasta'
seq_records = SeqIO.parse(open(file_in, mode='r'), 'fasta')
strains = []
dict_strain_seq = {}
for seq_record in seq_records:
    strains.append(seq_record.description)
    dict_strain_seq[seq_record.description]=str(seq_record.seq)
```

```python
n_strain = len(strains)
df = pd.DataFrame(columns=strains, index=strains)
n_strain = len(strains)
for i in range(n_strain):
    for j in range(0, i):
        seq_i = dict_strain_seq[strains[i]]
        seq_j = dict_strain_seq[strains[j]]
        if i != j:
            similarity = cal_similarity(seq_i, seq_j)
            df.iloc[i, j] =  similarity
            df.iloc[j, i] =  similarity
```
